```
import matplotlib.pyplot as plt
import numpy as np
```

```
lr = np.zeros(7)
it = range(7)
for i in range (7):
  if(i%2==1):
    lr[i] =1;

plt.xlabel('iterations')
plt.ylabel('lr')
plt.plot(it,lr)
```
